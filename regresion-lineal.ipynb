```python
n = 9
np.random.seed(77)
noise = np.random.normal(loc=0,scale=0.75,size=9)

xs = np.array(list(range(n)))-n//2
ys = np.array([1+x**2  for x in xs]) + noise
datos = np.vstack((xs,ys))
print(f"Datos:\n{datos}")

lr = Regresion_Lineal()
lr.fit(datos)

print(f"\na_0={lr.a_0}, a_1={lr.a_1}")

lr.plot()
lr.metrics()
```

```python
lr = Regresion_Lineal(grado=2)
lr.fit(datos)

print(f"\na_0={lr.a_0}, a_1={lr.a_1}, a_2={lr.a_2}")

lr.plot()
lr.metrics()
```
