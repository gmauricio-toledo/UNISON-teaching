```python
import numpy as np

text = '''
1
 2
 3.25
 4.5
 6
 7
 8
 9
 9.5
 10
x
 5
 6
 5.5
 7
 8.5
 8
 6
 7
 7
 5
'''

text = text.replace('\n','')
datos = text.split('x')
datos = np.array([[float(x) for x in datos[k].split()] for k in [0,1]])
datos = np.transpose(datos)
print(datos)
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi=150)
plt.scatter(datos[:,0],datos[:,1],color='red')
plt.xlabel("tiempo (s)",fontsize=14)
plt.ylabel("velocidad (m/s)",fontsize=14)
plt.xticks(range(int(np.max(datos[:,0]))+1))
plt.show()
```

```python
from interpolacion_aproximacion import RegresionLineal

rc = RegresionLineal(grado=2)
coeficientes = rc.fit(datos=datos)
rc.plot()
# rc.metrics()
```
