# Método de iteración de punto fijo

```python
import numpy as np
from math import sin, sqrt, e

def f(x):
    return 3-e**(-x)-x
    # return sin(sqrt(x))
    # return 2*x**3-11.7*x**2+17.7*x-5

def g(x):
    return f(x)+x

def error_relativo(x_real,x_aprox):
    return abs(x_real-x_aprox)/abs(x_real)

x_r = 1

tolerancia = 0.05

error = 1

aproximaciones = [x_r]
n_iteraciones = 0

while (error>tolerancia):
    x_r_old = x_r
    x_r = g(x_r)
    aproximaciones.append(x_r)
    if x_r != 0:
        error = error_relativo(x_r,x_r_old)
    n_iteraciones += 1

print("Método Punto Fijo ================")
print(f"Solución aproximada: {x_r}, Evaluación: {g(x_r)}")
print(f"Evaluación en f:{f(x_r)}")
print(f"Error relativo aproximado: {error}")
print(f"Iteraciones: {n_iteraciones}")
```

# Método de la secante

# Newton-Raphson

```python
import numpy as np
from math import sin, sqrt, e

def f(x):
    return 3-e**(-x)-x
    # return e**(-0.5*x)*(4-x)-2

def df(x):
    return e**(-x)-1
    # return (0.5*x - 3)*e**(-0.5*x)

def error_relativo(x_real,x_aprox):
    return abs(x_real-x_aprox)/abs(x_real)

x_r = 1
tolerancia = 0.05
error = 2*tolerancia

aproximaciones = [x_r]
n_iter = 0

print("\nMétodo Newton ===========")
while (error>tolerancia):
    x_r_old = x_r    
    x_r = x_r - f(x_r)/df(x_r)
    aproximaciones.append(x_r)
    if x_r != 0:
        error = error_relativo(x_r,x_r_old)
    print(f"Iteración: {n_iter+1} --- Aproximación: {x_r} --- Error: {error}")
    n_iter += 1

print(f"\nSolución aproximada: {x_r}, Evaluación: {f(x_r)}")
```
