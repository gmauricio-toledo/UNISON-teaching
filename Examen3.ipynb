```python
from edo import RungeKutta2, load_example
import numpy as np

example = load_example(3)

a = example['a']
b = example['b']
h = example['h']
print(f"[{a},{b}], h={h}")

solver = RungeKutta2(f=example['f'],
                    sol=example['sol'],
                    a2=1)
solver.fit(a,b,h,
            example['x0'],
            example['y0'])
solver.plot(plot_nodes=False)
solver.table()
```

```python
from edo import RungeKutta, load_example
import numpy as np

example = load_example(3)

a = example['a']
b = example['b']
h = example['h']
print(f"[{a},{b}], h={h}")

solver = RungeKutta(f=example['f'],
                    sol=example['sol'],
                    order=4
                    )
solver.fit(a,b,h,
            example['x0'],
            example['y0'])
solver.plot(plot_nodes=False)
solver.table()
```
