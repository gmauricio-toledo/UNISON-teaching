```python
from edo import RungeKutta, load_example
import numpy as np

example = load_example(3)

a = example['a']
b = example['b']
h = example['h']
print(f"[{a},{b}], h={h}")
print(example['function_f'])

solver = RungeKutta(f=example['f'],
                    sol=example['sol'],name='Heun')
solver.fit(a,b,h,
            example['x0'],
            example['y0'])
solver.plot(plot_nodes=False)
solver.table()
```
