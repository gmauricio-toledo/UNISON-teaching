```python
import pandas as pd

df = pd.read_excel('Calificaciones AS-LCC.xlsx',sheet_name='Nombres')
df
```

```python
import numpy as np
from itertools import combinations
from random import shuffle, sample

n = 12 # Ejercicios totales
N = df.shape[0] # Los que presentarán
num_ejercicios = 4

ejercicios = range(1,n+1)
asignaciones = [x for x in combinations(ejercicios,num_ejercicios)]
idxs = [j for j,x in enumerate(asignaciones)]
lista = np.random.choice(idxs,size=N,replace=False)
asignacion_final = [sample(asignaciones[j],k=num_ejercicios) for j in lista]
print(asignacion_final)
```

```python
for j,X in enumerate(asignacion_final):
    lista_ejercicios = str(X[0])
    for x in X[1:]:
        lista_ejercicios += ", " + str(x)
    df.loc[j,'Ejercicios'] = lista_ejercicios
df
```

```python
def dataframe_to_latex_table(df):
    '''
    This function receives a pandas dataframe and returns the latex code for a 
    table containing the information stored in the dataframe.
    '''
    num_cols = df.shape[1]
    num_columns_str = '|' + num_cols*'c|'
    table_body_script = ""
    for j in df.index.to_list():
        row = str(df.iloc[j,0])
        for col_name in df.columns.to_list()[1:]:
            row += " & " + str(df.loc[j,col_name])
        row += "\\\\ \\hline"
        table_body_script += row + "\n" 
    script = f'''\\begin{{table}}[h]
        \\centering
        \\begin{{tabular}}{{{num_columns_str}}}
            {table_body_script}
        \\end{{tabular}}
    \\end{{table}}'''
    return script
```

```python
script = dataframe_to_latex_table(df)
print(script)
```
