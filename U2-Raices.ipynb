```python
from numpy import e
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return e**(-x)+x-2

# def f(x):
#     return x**10 - 1
```

Graficamos la funci√≥n. Usamos [numpy.linspace](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html) para generar los puntos sobre el eje X.

```python
eje_x = np.linspace(start=-2.5,
                    stop=2.5,
                    num=100)

# eje_y = [f(x) for x in eje_x]  # Listas de compresi√≥n 

eje_y = []
for x in eje_x:
    eje_y.append(f(x))

plt.figure(figsize=(7,5))
plt.plot(eje_x,eje_y,color='red')
plt.axhline(0,color='gray') # Dibujar eje X
plt.axvline(0,color='gray') # Dibujar eje Y
plt.show()
```

# Punto fijo

```python
from math import e
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return e**(-x)-x
```

```python
eje_x = np.linspace(start=-2.5,
                    stop=2.5,
                    num=100)

# eje_y = [f(x) for x in eje_x]  # Listas de compresi√≥n 

eje_y = []
for x in eje_x:
    eje_y.append(f(x))

plt.figure(figsize=(7,5))
plt.plot(eje_x,eje_y,color='red')
plt.axhline(0,color='gray') # Dibujar eje X
plt.axvline(0,color='gray') # Dibujar eje Y
plt.show()
```

```python
eje_x = np.linspace(start=-2.5,
                    stop=2.5,
                    num=100)

eje_y = [f(x)+x for x in eje_x]  # Listas de compresi√≥n 

plt.figure(figsize=(7,5),dpi=150)
plt.suptitle("Iteraci√≥n de punto fijo")
plt.plot(eje_x,eje_y,color='red')
plt.plot(eje_x,eje_x,color='black')
plt.axhline(0,color='gray') # Dibujar eje X
plt.axvline(0,color='gray') # Dibujar eje Y
plt.show()
```

```python
from roots import RootFinder

rf = RootFinder()
rf.fit(f=f,x0=0,tolerance=0.001,method='fix')
rf.print_table(digits=7)
```

```python
rf.animate(fname='ex3-fix.gif')
```

El siguiente ejemplo siempre diverge con el m√©todo del punto fijo. **Cortar la ejecuci√≥n en alg√∫n momento, de otra forma se queda ejecut√°ndose indefinidamente**

```python
def f2(x):
    return x**3-x-1

rf = RootFinder()
rf.fit(f=f2,x0=-3,tolerance=0.001,method='fix')
rf.print_table(digits=7)
```

# Generar animaciones

```python
import itertools
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from math import fabs

data_x = itertools.cycle(rf.roots)
data_y = itertools.cycle(np.zeros_like(rf.roots))
frames = len(rf.roots)

window_width = fabs(rf.x1-rf.x0)

eje_x = np.linspace(start=rf.x0-0.1*window_width,
                    stop=rf.x1+0.1*window_width,
                    num=100)
eje_y = [f(x) for x in eje_x] 

fig, ax = plt.subplots(figsize=(7,5))

def update(i):
    ax.clear()
    fig.suptitle(f"Iteraci√≥n {i}")
    ax.axhline(0,color='gray')
    ax.plot(eje_x,eje_y,color='red')
    ax.scatter([next(data_x)], [next(data_y)],s=50,color='black')


ani = animation.FuncAnimation(fig, update, frames=frames, interval=500)
ani.save('ex1-newton.gif', writer='pillow')
```

# Casos problem√°ticos

```python
from math import sqrt

def f(x):
    if x>=0:
        y = sqrt(x)
    else:
        y = -sqrt(-x)
    return y

def df(x):
    if x>=0:
        y = 1/(2*sqrt(x))
    else:
        y = 1/(2*sqrt(-x))
    return y   
```

```python
from roots import RootFinder

rf = RootFinder()
rf.fit(f=f,df=df,x0=0.25,tolerance=0.000001,method='newton')
```
